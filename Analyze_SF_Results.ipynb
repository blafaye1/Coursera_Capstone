```python
import numpy as np
import pandas as pd
import requests
import json
import folium
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
import time
```

Load hardcoded variables and secrets

```python
with open('variables/city_centers.json') as f_in:
    city_centers = json.load(f_in)

sf_lat = city_centers['sf'][0]
sf_lon = city_centers['sf'][1]
chicago_lat = city_centers['chicago'][0]
chicago_lon = city_centers['chicago'][1]
nyc_lat = city_centers['nyc'][0]
nyc_lon = city_centers['nyc'][1]
```

Load data

```python
with open('json_data/sf_venues_results.json') as f_in: 
    sf_venues_results = json.load(f_in)
```

Analysis

```python
len(sf_venues_results)
```

```python
sf_venues_map = folium.Map(location = [sf_lat, sf_lon], zoom_start = 11)

unique_ids = set()
for i, venue in enumerate(sf_venues_results):
    items_list = venue['response']['groups'][0]['items']
    for item in items_list:
        item_id = item['venue']['id']
        if item_id in unique_ids:
            continue
        else:
            unique_ids.add(item_id)
            item_lat = item['venue']['location']['lat']
            item_lon = item['venue']['location']['lng']
            item_cat = item['venue']['categories'][0]['name']
            
            folium.CircleMarker(
                [item_lat, item_lon],
                radius = 5,
                popup = item_cat,
                color = 'blue',
                fill = True).add_to(sf_venues_map)

sf_venues_map
```
